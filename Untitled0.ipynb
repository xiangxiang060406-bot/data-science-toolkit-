```python
import numpy as np
# Create2x2numpyarray
arr_2x2 = np.array([[5, 6], [7, 8]])
print("1. 2x2 numpy array:")
print(arr_2x2)
```

```python
# Create a 3x4 random numpy array (with elements between 0 and 1)
random_3x4 = np.random.rand(3, 4)
print("\n2. 3X4 rando mnumpy array:")
print(random_3x4)
```

```python
# Create an array of all 1s
ones_array = np.ones((3, 2))
# Create an array of all zeros
zeros_array = np.zeros((2, 4))
print("\ N3. array of all 1s:")
print(ones_array)
print("Array of all zeros:")
print(zeros_array)
```

```python
#Create a 6x3 random array
random_6x3 = np.random.rand(6, 3)
#Transpose array (shape becomes 3x6)
transposed_array = random_6x3.T
print("\n4. Shape of a 6x3 random array:", random_6x3.shape)
print("Transposed array shape:", transposed_array.shape)
print("After transposing the array:")
print(transposed_array)
```

```python
#Define dimensions (customizable, here take n=2, p=5, m=3 as an example)
n, p, m = 2, 5, 3
#Create a random array of n×p and p×m
arr_np = np.random.rand(n, p)
arr_pm = np.random.rand(p, m)
#Compute the dot product (the result is nxm)
dot_product = np.dot(arr_np, arr_pm)
print("\ N5. n×p array shape:", arr_np.shape)
print("P×m array shape:", arr_pm.shape)
print("Dot product result shape:", dot_product.shape)
print("Dot product result:")
print(dot_product)
```
